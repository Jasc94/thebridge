```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
def matrix_creator():
    # Board's index
    index = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
    # Two dicts for later use
    d = {}
    d2 = {}

    # The outer dict is for the indexes and the inner for the columns
    for i in index:
        for j in range(1, len(index) + 1):
            # I fill it with '~'
            d2[j] = 1
        d[i] = d2

    # Dataframe out of the dicts
    df = pd.DataFrame(d)
    # Transpose it to have the letters to be the index and the numbers the columns
    df = df.T
    return df
```

```python
df = matrix_creator()

plt.figure(figsize=(15,10))
sns.heatmap(df, cmap="BrBG", center = 0, linewidths = 0.5, linecolor = "white")
```

```python
def matrix_creator():
    # Board's index
    index = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
    # Two dicts for later use
    d = {}
    d2 = {}

    # The outer dict is for the indexes and the inner for the columns
    for i in index:
        for j in range(1, len(index) + 1):
            # I fill it with '~'
            d2[j] = '~'
        d[i] = d2

    # Dataframe out of the dicts
    df = pd.DataFrame(d)
    # Transpose it to have the letters to be the index and the numbers the columns
    df = df.T
    return df
```

```python
df = matrix_creator()
df.loc["A", 1:9] = '*'
df.loc["B", 1:9] = '*'
df.loc["C", 4:8] = 'X'
df.loc["D", 4:8] = 'O'
df
```

```python
new_df = df.replace('*', 1)
new_df
```

```python
new_df2 = df.replace(['~', '*', 'O', 'X'], [1, 1, 0, -1])
new_df2
```

```python
front_board_creator(new_df2)
```
