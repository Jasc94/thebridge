### Step 1. Import the necessary libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# print the graphs in the notebook
%matplotlib inline

# set seaborn style to white
sns.set_style("white")
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

```python
tips = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv")
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop("Unnamed: 0", axis = 1, inplace = True)
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
sns.distplot(tips["total_bill"])

plt.xlabel("Value")
plt.ylabel("Frequency")

plt.title("Total bill")
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "scatter")
```

### Step 7.  Create one image with the relationship of `total_bill`, `tip` and `size`.
#### Hint: It is just one function.

```python
sns.pairplot(tips, kind = "scatter")
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.catplot(x = "day", y = "total_bill", data = tips)

plt.grid(axis = "y")
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.catplot(x = "total_bill", y = "day", hue = "sex", data = tips)

plt.grid(axis = "x")
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.catplot(x="day", y="total_bill", hue="time", kind="box", data=tips, width = 0.8, aspect = 1.5, legend_out = False)

plt.legend(loc = 1)
plt.grid(axis = "y")
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True)

sns.histplot(data = tips[tips["time"] == "Dinner"], x = "tip", bins = 10, ax = ax1)
sns.histplot(data = tips[tips["time"] == "Lunch"], x = "tip", bins = 10, ax = ax2)

ax1.set_title("time = Dinner")
ax2.set_title("time = Lunch")

plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize = (8, 6))

sns.scatterplot(x = "total_bill", y = "tip", hue = "smoker", data = tips[tips["sex"] == "Female"], ax = ax1)
sns.scatterplot(x = "total_bill", y = "tip", hue = "smoker", data = tips[tips["sex"] == "Male"], ax = ax2, legend = False)

ax1.set_title("Female")
ax2.set_title("Male")

plt.show()
```
