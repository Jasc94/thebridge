### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Step 2. Create the DataFrame it should look like below.

This time you will create the data.

```python
first_name = ["Jason", "Molly", "Tina", "Jake", "Amy"]
last_name = ["Miller", "Jacobson", "Ali", "Milner", "Cooze"]
age = [42, 52, 36, 24, 73]
female = [0, 1, 1, 0, 1]
preTestScore = [4, 24, 31, 2, 3]
postTestScore = [25, 94, 57, 62, 70]
```

```python
s1 = pd.Series(first_name, name = "first_name")
s2 = pd.Series(last_name, name = "last_name")
s3 = pd.Series(age, name = "age")
s4 = pd.Series(female, name = "female")
s5 = pd.Series(preTestScore, name = "preTestScore")
s6 = pd.Series(postTestScore, name = "postTestScore")
```

```python
df = pd.concat([s1, s2, s3, s4, s5, s6], axis = 1)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(df["preTestScore"], df["postTestScore"], s = df["age"])

plt.xlabel("preTestScore")
plt.ylabel("postTestScore")

plt.title("preTestScore vs postTestScore by Age")
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.scatter(df["preTestScore"], df["postTestScore"], s = 4 * df["postTestScore"], c = df["female"], cmap = "RdYlBu_r")

plt.xlabel("preTestScore")
plt.ylabel("postTestScore")

plt.title("preTestScore vs postTestScore by Age")
```
