Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  
  >Enter a number:

  >> 1234

  >>> 2 cows, 0 bulls

  >> 1253

  >>> 1 cow, 1 bull
  
  ...
Until the user guesses the number or write "STOP"

```python
# -------- Modules --------
import random
from functools import reduce

# -------- Function to check number of cows/bulls in guess --------
def checker(num, guess):
    num = str(num)
    guess = str(guess)

    cows = 0    # Value and position
    bulls = 0   # Position

    for i, x in enumerate(num):
        if x in guess:
            bulls += 1

            if guess[i] == x:
                bulls -= 1
                cows += 1

    return [bulls, cows]

# -------- To create the random number to guess --------
def random_non_repetitive():
    x = []
    count = 0

    while count < 4:
        nr = random.randint(0, 9)
        if nr not in x:
            x.append(str(nr))
        count += 1

    num = int(reduce(lambda x, y: x + y, x))
    return num

# -------- To play the game --------
# I have to pass it a num. In case I want to have a random number, then I pass the function above
def game(num):
    bulls, cows = [0, 0]

    while cows < 4:
        guess = input('Enter your guess: ')
        if guess == 'STOP':
            break

        print('Your guess:', guess)
        bulls, cows = checker(num, guess)
        print('{} cows, {} bulls'.format(cows, bulls))

    print('Your guess: ', guess)
    print('Your num: ', num)
    print('\nOle tu polla. Lo has adivinado.')
```
