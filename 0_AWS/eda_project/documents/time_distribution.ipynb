```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns

fig = plt.figure(figsize = (8, 8))
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')

tasks = ['Find the project', 'Find the data related to the project', 'Define a hypothesis', 'Define the necessary steps to demonstrate the hypothesis', 'Work with data + coding', "Document all the steps and send it to teachers"]
time = [1, 2, 1, 1, 4, 2]

color_map = sns.color_palette("Set2")

ax.pie(time, labels = tasks, autopct = '%1.2f%%', colors = color_map)

plt.show()
plt.savefig("Time_distribution.png")
```
